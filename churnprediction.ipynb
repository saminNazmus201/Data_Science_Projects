```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test= train_test_split(x , y,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
x_train= sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=  LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)
```

```python
ypred=lr.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
```

```python
from sklearn.pipeline import Pipeline
```

```python
new_param_grid = [
    {
        'logreg__solver': ['liblinear'],
        'logreg__penalty': ['l1', 'l2'],
        'logreg__C': [0.01, 0.1, 1, 10],
        'logreg__class_weight': [None, 'balanced']
    },
    {
        'logreg__solver': ['saga'],
        'logreg__penalty': ['elasticnet'],
        'logreg__l1_ratio': [0.1, 0.5, 0.9],
        'logreg__C': [0.01, 0.1, 1, 10],
        'logreg__class_weight': [None, 'balanced']
    }
]
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=2000))
])


inner_cv = KFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    estimator=pipe, 
    param_grid=new_param_grid, 
    cv=inner_cv, 
    scoring='f1'  
)


outer_cv = KFold(n_splits=3, shuffle=True, random_state=42)

nested_scores = cross_val_score(grid_search, x, y, cv=outer_cv, scoring='f1')


print(f"F1-Scores for each outer fold: {nested_scores}")
print(f"Mean Nested CV F1-Score: {nested_scores.mean():.4f}")
```

```python
param_grid = [
    {
        # Liblinear supports L1 and L2
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [0.01, 0.1, 1, 10, 100],
        'class_weight': [None, 'balanced']
    },
    {
        # SAGA is the ONLY one that supports ElasticNet
        'solver': ['saga'],
        'penalty': ['elasticnet'],
        'l1_ratio': [0.1, 0.5, 0.9],
        'C': [0.01, 0.1, 1, 10, 100],
        'class_weight': [None, 'balanced']
    },
    {
        # LBFGS only supports L2
        'solver': ['lbfgs'],
        'penalty': ['l2'],
        'C': [0.01, 0.1, 1, 10, 100],
        'class_weight': [None, 'balanced']
    }
]
log_reg = LogisticRegression(max_iter=2000)
grid_search = GridSearchCV(log_reg, param_grid, scoring='recall_macro', cv=5)
grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(x_test)

accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
conf_matrix_tuned = confusion_matrix(y_test, y_pred_tuned)

print("Accuracy:", accuracy_tuned)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x=df.drop(columns=["Churn"])
y=df["Churn"]
```

```python
x_train , x_test , y_train , y_test= train_test_split(x , y,test_size=0.2, random_state=42)
```

```python
dtcl=DecisionTreeClassifier()
dtcl.fit(x_train, y_train)
```

```python
dtypred= dtcl.predict(x_test)
dtacc= accuracy_score(y_test, dtypred)
print(dtacc)
dtf1= f1_score(y_test, dtypred)
print(dtf1)
```

```python
paramDT=[{
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced']
}]
gridsearchDT= GridSearchCV(DecisionTreeClassifier(), paramDT, cv=5, scoring='f1', n_jobs=-1)
gridsearchDT.fit(x_train, y_train)
bestDT= gridsearchDT.best_estimator_
tunedDTy= bestDT.predict(x_test)
```
